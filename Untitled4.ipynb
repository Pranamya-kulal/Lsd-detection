```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create an instance of ImageDataGenerator for data augmentation
datagen = ImageDataGenerator(rescale=1.0/255.0)  # Normalize pixel values to [0, 1]

# Load images from the directory structure
base_dir = '/content/dataset/train'  # Base directory for training data

data_generator = datagen.flow_from_directory(
    base_dir,  # Use the parent directory
    target_size=(150, 150),  # Resize images to 150x150
    class_mode='binary',  # Set the class mode to binary
    batch_size=32,  # Set the batch size
    shuffle=True  # Shuffle the data
)

# Check the number of batches
num_batches = len(data_generator)
print(f"Number of batches: {num_batches}")
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define the model
model = Sequential()

# First convolutional layer
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Second convolutional layer
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Third convolutional layer
model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the output
model.add(Flatten())

# Fully connected layer
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # Dropout for regularization

# Output layer
model.add(Dense(1, activation='sigmoid'))  # Binary classification

# Summary of the model
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(data_generator,
                    steps_per_epoch=num_batches,
                    epochs=10)  # You can adjust the number of epochs as needed
```

```python
import os
import shutil
import numpy as np

# Define the paths
train_data_dir = '/content/dataset/train'  # Adjust this path if necessary
val_data_dir = '/content/dataset/validation'  # Path for validation dataset

# Create the validation directory if it doesn't exist
if not os.path.exists(val_data_dir):
    os.makedirs(val_data_dir)
    os.makedirs(os.path.join(val_data_dir, 'healthy'))
    os.makedirs(os.path.join(val_data_dir, 'unhealthy'))

# Split the training images into training and validation sets
for category in ['healthy', 'unhealthy']:
    category_path = os.path.join(train_data_dir, category)
    images = os.listdir(category_path)
    np.random.shuffle(images)

    # Get 20% of the images for validation
    num_val_images = int(0.2 * len(images))
    val_images = images[:num_val_images]

    # Move validation images to the validation directory
    for img in val_images:
        shutil.move(os.path.join(category_path, img), os.path.join(val_data_dir, category, img))

print("Validation dataset created successfully!")
```

```python
import tensorflow as tf # If not already imported

def generate_data_with_weights(generator, class_weights):
  """
  Wraps the original generator to include sample weights.

  Args:
    generator: The original data generator.
    class_weights: A dictionary mapping class indices to weights.

  Yields:
    A tuple of (image, label, weight).
  """
  for image, label in generator:
    weight = tf.gather(tf.constant(list(class_weights.values()), dtype=tf.float32), tf.cast(label, tf.int32))
    # If label is a one-hot encoded vector:
    # weight = tf.reduce_sum(tf.constant(list(class_weights.values()), dtype=tf.float32) * label, axis=-1)
    yield image, label, weight

# Update the fit method to use the modified generator:
class_weights = {0: 1, 1: 5}  # Adjust these weights based on class distribution
train_data_generator_with_weights = generate_data_with_weights(train_data_generator, class_weights)
model.fit(train_data_generator_with_weights, class_weight=None) # class_weight is now handled by the
```
